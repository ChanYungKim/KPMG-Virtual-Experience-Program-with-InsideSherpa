```
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

from google.colab import drive

drive.mount('/content/gdrive/')
NewCustomers = pd.read_csv('/content/gdrive/My Drive/KPMG/NewCustomers.csv')
```

```
NewCustomers = NewCustomers[['first_name', 'last_name', 'gender','DOB', 'job_industry_category', 'wealth_segment', 'owns_car', 'state']]
```

```
NewCustomers.head()
```

```
NewCustomers['DOB'] = pd.to_datetime(NewCustomers.DOB)

NewCustomers['AGE'] = ((pd.datetime.datetime.now()- NewCustomers.DOB)).dt.days

NewCustomers['AGE'] = np.floor(NewCustomers.AGE / 365)
NewCustomers = NewCustomers[NewCustomers.AGE <= 100]

age_bucket = [0, 20, 40, 60, 80, 100]
age_groups = ['Age : 0 - 20', 'Age : 21 - 40', 'Age : 41 - 60', 'Age : 61 - 80', 'Age : 81 - 100']
NewCustomers['Age_Groups'] = pd.cut(NewCustomers['AGE'], age_bucket, labels=age_groups)

NewCustomers.dropna(subset = ['job_industry_category'], how = 'any', inplace = True)
NewCustomers.isna().sum()
```

```
print (NewCustomers.job_industry_category.unique())
print (NewCustomers.gender.unique())
print (NewCustomers.state.unique())
print (NewCustomers.wealth_segment.unique())
print (NewCustomers.owns_car.unique())
```
