```
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

from google.colab import drive

drive.mount('/content/gdrive/')
data = pd.read_csv('/content/gdrive/My Drive/KPMG/Transactions.csv', index_col='transaction_date')

data.head()
data.info()
```

```
#data.standard_cost = data.standard_cost.astype(str)
data.standard_cost = data.standard_cost.str.replace('$', '')
data.standard_cost = data.standard_cost.str.replace(',', '').astype(float)

data.standard_cost.head()
```

```
data = data[['brand', 'list_price', 'standard_cost']]

data['profit'] = (data.list_price - data.standard_cost)

data.index = pd.to_datetime(data.index)

data.sort_values('transaction_date', inplace = True)

data.dropna(how = 'any', inplace = True)

data.head()
```

```
Brands_Sold = data.groupby('transaction_date').brand.value_counts(sort = False)
Brands_Sold.to_excel('/content/gdrive/My Drive/KPMG/Brands_Sold.xlsx')
Brands_Sold.head()
```

```
Brands_Profit = data.groupby(['transaction_date', 'brand']).profit.sum()
Brands_Profit.to_excel('/content/gdrive/My Drive/KPMG/Brands_Profit.xlsx')
Brands_Profit.head()
```

```
data['date'] = data.index
data['month'] = data.date.dt.month.astype(int)

data.head()
data.tail()
```

```
data_monthly = data[['month', 'brand', 'profit']]

Brands_Sold_Monthly = data_monthly.groupby(['brand', 'month']).brand.value_counts()
Brands_Sold_Monthly.to_excel('/content/gdrive/My Drive/KPMG/Brands_Sold_Monthly.xlsx')
Brands_Sold_Monthly.head()
```

```
Brands_Profit_Monthly = data.groupby(['brand', 'month']).profit.sum()
Brands_Profit_Monthly.to_excel('/content/gdrive/My Drive/KPMG/Brands_Profit_Monthly.xlsx')
Brands_Profit_Monthly.head()
```
