```
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

from google.colab import drive

drive.mount('/content/gdrive/')
my_data = pd.read_csv('/content/gdrive/My Drive/KPMG/Training_Dataset.csv', index_col='customer_id')
```

```
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score
```

```
X = my_data.drop(columns= ['Customer_Class'])
y = my_data.Customer_Class

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

column_transform = make_column_transformer((OneHotEncoder(), ['Age_Groups', 'gender', 'job_industry_category', 'state', 'owns_car', 'wealth_segment']))
```

```
from sklearn.linear_model import LogisticRegression

LR_classifier = LogisticRegression(solver = 'lbfgs')

pipe = make_pipeline(column_transform, LR_classifier)

cross_val_score (pipe, X_train, y_train, cv = 5, scoring = 'accuracy').mean()
```

```
pipe.fit(X_train, y_train)

predictions = pipe.predict(X_test)

print(accuracy_score(y_test, predictions))
```

```
NewCustomers = pd.read_csv('/content/gdrive/My Drive/KPMG/NewCustomers_Dataset.csv', index_col=['first_name', 'last_name'])
NewCustomers = NewCustomers [['gender', 'job_industry_category', 'wealth_segment', 'owns_car', 'state', 'Age_Groups']]

predictions = pipe.predict(NewCustomers)

NewCustomers['Customer_Class'] = predictions

NewCustomers.sort_index(inplace = True)

NewCustomers.head()
```
